## 78: Find reasonForVisit with highest count of acuity 5 patients. 

```python
df= pd.read_excel('/Users/nishka/Downloads/HospitalDatabase.xlsx', sheet_name='EDVisits')
# Create a new DataFrame with only acuity 5 patients
df_acuity5 = df[df['Acuity'] == 5]

# Group the new DataFrame by reason for visit and count the number of patients in each group
counts = df_acuity5.groupby('ReasonForVisit').size()

# Get the reason for visit with the highest count
most_common_reason = counts.idxmax()

print("The most common reason for visit among acuity 5 patients is:", most_common_reason)
```
